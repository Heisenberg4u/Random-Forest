```python
# To make data visualisations display in Jupyter Notebooks
import pandas as pd     # Data processing, Input & Output load
import numpy as np      # linear algebra
import seaborn as sns    # statistical visuaization
import matplotlib.pyplot as plt    # Visualization & plotting
%matplotlib inline
import warnings   # To avoid warning messages in the code run
warnings.filterwarnings("ignore")
```

```python
employee = pd.read_csv('Employees data.csv')
```

```python
employee.head()
```

```python
#We want  data to build a model to predict in future who will leave the company & who will not leave the company .
#Here attrition means left
```

```python
employee.describe()
```

```python
#^It has described all the numerical variables.Here std. deviation=0 means there is no deviation 
#in the values. Every row has same value for that column so employee count column m  1,there is 
# ek hi value h i.e no variation in the value.
```

```python
employee.nunique()
```

```python
# ^This will tell us no. of unique values in every column,so we want 2 values atleast in every column where ever the value is 1
# like 'Employeecount','Over18','Standard Hours' these are completely useless they are not adding anything hence rather than
#  handling so much data we can  remove these columns.
```

```python
employee = employee.drop(columns=['EmployeeCount','Over18','StandardHours'])
```

```python
employee.nunique()     #these 3 columns are gone now
```

```python
employee.head()
```

```python
#^infact employee Number is useless its like a serial number or Id,this will not have any impact on the attrition so we can
#drop this also (below)
```

```python
employee = employee.drop(columns=['EmployeeNumber'])
```

```python
employee.isnull().sum()    #to see if we have any missing values or not
```

```python
employee.head()
```

```python
#^O/P= We don't have any missing values &  we don't have to do handling from decision trees as Decision Trees are 
# insensitive to outliers
```

```python
#The key variable which we want to predict is attrition let's see how many values is it making :
```

```python
employee['Attrition'].value_counts()   #or we can see through nunique too
```

```python
#^O/P= 'No' means employee hasn't left the company & yes means he left .It means most of the Employees have not left the company
```

```python
#Now we will change the attrition values into 0 & 1
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
employee['Attrition'] = labelencoder.fit_transform(employee['Attrition'])
```

```python
employee.head()
```

```python
#^O/P=Now it has changed to 0(No) & 1(Yes).Now it is our Y rest all is X
```

```python
employee['Attrition'].value_counts()    #To see kisko 0 diya h and kisko 1.there 1 is Yes and 0 is 
```

```python
#Now we need to convert these categorical/char. variables like 'Business Travel'&'Department' values into dummy variables
#basically we need to separate numerical data & character data & then char. data ko dummy variable m change krenge
```

```python
#let's separate the numerical data
employee_num = employee.select_dtypes(include=[np.number])
```

```python
employee_num.head()
```

```python
#Similarly let's separate character data
employee_chararcter = employee.select_dtypes(include='object')
```

```python
employee_chararcter.head()
```

```python
#Now we will convert them into dummy variables
employee_dummies = pd.get_dummies(employee_chararcter,drop_first=True)
```

```python
employee_dummies.head()
```

```python
#^O/P=1) We have converted Character data into dummy numerical values .2) Here drop_first=true means that we dont have to create
# two different columns for females instead we can manage in one column only  in Gender_ Male Column 0 stands for female and 1 
# for male.
```

```python
employee_combined = pd.concat([employee_num,employee_dummies],axis=1)
```

```python
employee_combined.shape
```

```python
#Now we will split the data in Independent variables(X) and Dependent variables(Y)
```

```python
X = employee_combined.drop(columns=['Attrition'],axis=1)
```

```python
Y = employee_combined['Attrition']
```

```python
X.shape, Y.shape
```

```python
#Now we will split the data into training & test data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,train_size=0.7, random_state=20)
```

```python
X_train.shape, X_test.shape, Y_train.shape, Y_test.shape
```

# Case 1: Underfitted model

```python
from sklearn import tree, metrics

dt_model = tree.DecisionTreeClassifier(criterion='entropy',
                                      max_depth=1,
                                      random_state=20)
dt_model.fit(X_train,Y_train)

Y_pred_train = dt_model.predict(X_train)
Y_pred_test = dt_model.predict(X_test)

print("Accuracy Train:",metrics.accuracy_score(Y_train,Y_pred_train))
print("Accuracy Test:",metrics.accuracy_score(Y_test,Y_pred_test))

metrics.confusion_matrix(Y_test,Y_pred_test)
```

```python
Y_pred_test
```

```python
#^ Here max_depth=1 because we are just taking example of underfitted model,to reduce the complexity.
```

```python
#^O/P= If we are using the data of Attrition or employees then every time Attrition  is predicting 'No' only.This is happening 
# because our  maximum data is of 'No' type.Ideally we should ideally split the data equally/proportionaly in 0 and 1 or we can
#  do class_weight='balanced' 

#Ideally we should prefer to split the data equally/proportionaly in 0 and 1 rather than doing  class_weight='balanced'
```

```python
from sklearn import tree, metrics

dt_model = tree.DecisionTreeClassifier(criterion='entropy',
                                      max_depth=1,
                                       class_weight='balanced',
                                      random_state=20)
dt_model.fit(X_train,Y_train)

Y_pred_train = dt_model.predict(X_train)
Y_pred_test = dt_model.predict(X_test)

print("Accuracy Train:",metrics.accuracy_score(Y_train,Y_pred_train))
print("Accuracy Test:",metrics.accuracy_score(Y_test,Y_pred_test))

metrics.confusion_matrix(Y_test,Y_pred_test)
```

```python
Y_pred_test
```

# Random Search CV

```python
#It is use to tune the hyperparameters of various algorithms.We have to tell  machine to try these values and among
# those values machine will choose the best value.
```

```python
tree.DecisionTreeClassifier()
```

```python
#^O/P=these are the hyperparameters of DT,which can pick any of them & try to tune it
```

```python
model_params = {"criterion":['gini','entropy'],
                "max_depth":range(1,10),
                "max_features":range(5,15)}
```

```python
#O/P=We have to tell this to machine and ye bhi ki hyperparameters konsi algorithm m try krna h
```

```python
2*9*9
```

```python
#^O/P=1)We can have 162 different combinations from these 3 hyperparameters.
#2)If we have to ask what's the best out of these possible values,we need to try 162 different combinations & out of it the one
# with the best combination  will be best for us but this is not practically possible (manually to do this) & for this an 
# inbuilt loop in Random Search Cv will try different combinations and will give the best answer among all combinations.
```

```python
#Now we will implement Random Search CV,first we are trying from decision tree.ye algo_type we have to mention
from sklearn.model_selection import RandomizedSearchCV
algo_type = tree.DecisionTreeClassifier(random_state=15)
```

```python
#Now we can build our model
models = RandomizedSearchCV(algo_type,model_params,random_state=15)
```

```python
#^O/P=Here we have told the  machine to tune the hyperparameters in DTclassifier i.e(algo_type,model_params,random_state=15)
```

```python
#^NOTE=1)RandomizedSearchCV k andar ek hyperparameter h n_iteration which is by default 10(tab+shift shift),this means humare
#162 model_params k jo humne pass kra h RandomizedSearchCV m
```

```python
#2)Now RandomizedSearchCV will not tryout all the possible 162 different combinations ,wo sirf 10 combinations ko tryout krega 
#and those combinations would be searched randomly so maybe the first combination can be 'gini','max_depth=5',& 'max_feat=2'.
#next combination can be 'entropy','max_depth=',& 'max_feat=12',like this it will try out those 10 different combinations out
#162 iff. combns. and from  thise 10 different combinations it will give us the best combination which is giving us the best
#Result so that means wo konse combinations tryout krga usse randomness aa skti h isliye to avoid randomness agr hum random_stat
#e=15 (ya  any same value) mention kre toh same combination tryout krega.agar hum ranom_state same mention nhi krenge toh it 
#will work but will give diff. result(person to person) & we can give n_iter=100 toh wo ab 162 m se 100 different combination 
#tryout krega so there are chances of getting better result although utna zyada time lgega,utni computational power lagegi.
#So if we have enough  time,enough computational power then we can try out a lot of diff. combos it will result in the best o/p.
```

```python
models = RandomizedSearchCV(algo_type,model_params,n_iter=100,random_state=15)
```

```python
#^O/P=It has just created dummy combinations as of now unn combinations ko try nhi kra h qki humne data hi nhi diya,try k liye
```

```python
rscv_best_model = models.fit(X_train,Y_train)
```

# Grid Search CV

```python
#In RandomSEarch CV we were mentioning n_itration & it was trying just that many combinations but Grid Search CV m jitne bhi 
#Different combinations bn skte h humare hyperparameters k basis pr ye utne combinations try krta h,ByDefault sab try krta h
#It will take a lot of time to run though but it does equal job or better job.
```

```python
model_params = {"n_estimators":[80,90,100,110,120,130,140,150,160],
               "criterion":['gini','entropy'],
               "max_depth":range(3,8)}

from sklearn.model_selection import GridSearchCV
algo_type = ensemble.RandomForestClassifier(random_state=15)

models = GridSearchCV(algo_type,model_params)

rscv_best_model = models.fit(X_train,Y_train)

Y_pred_train = rscv_best_model.predict(X_train)
Y_pred_test = rscv_best_model.predict(X_test)

print("Accuracy Train:",metrics.accuracy_score(Y_train,Y_pred_train))
print("Accuracy Test:",metrics.accuracy_score(Y_test,Y_pred_test))

print(rscv_best_model.best_params_)
```
